```python
import pandas as pd
```

```python
po_df = pd.read_csv('amazonLTLPO.csv')
po_df.head()
```

```python
#clean up data

new_df = po_df[['Order/PO Number', 'ASIN', 'Model Number', 'List Price', 'Discount', 'Cost', 'Quantity Ordered', 'Quantity Confirmed', 'Quantity Received']]
new_df["List Price"] = new_df["List Price"].replace('[\$,]', '', regex=True).astype(float)
new_df["Cost"] = new_df["Cost"].replace('[\$,]', '', regex=True).astype(float)
new_df['Total'] = new_df['List Price'] * new_df['Quantity Ordered']
new_df
```

```python
#get the top ten items

# i define top ten is by quanttity ordered or total price?

#first see by quantity ordered

topQuantity = new_df.groupby('Model Number').sum().sort_values('Quantity Ordered', ascending = False )
topQuantity
```

```python
topQuantity.index[0:10]
```

```python
#now by total money generated 

topTotal = new_df.groupby('Model Number').sum().sort_values('Total', ascending = False )
topTotal
```
